```
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sn
```

```
X = pd.read_csv("/combined_data_no-states.csv", sep=",", header = 1)
corrMatrix = X.corr()
corrMatrix.to_excel("corrMatrix.xlsx")
```

```
sn.heatmap(corrMatrix, annot=False, xticklabels=True, yticklabels = True)
figure = plt.gcf()
figure.set_size_inches(20, 20)
plt.savefig("corrMatrix", dpi=100)
plt.show()
```
