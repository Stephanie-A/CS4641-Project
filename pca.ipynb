```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read csv input file
X = pd.read_csv("combined-data-new-no-header.csv", sep=",")
```

```python
for i in range(min(X.shape[0], X.shape[1])):
    pca = PCA(n_components=i)
    pca.fit(X)
    print(pca.singular_values_)
    
    plt.plot([*range(len(pca.singular_values_))], pca.singular_values_, 'bx-')
    plt.xlabel('Singular Values')
    plt.ylabel('Variance')
    title  = 'PCA ' + str(i);
    plt.title(title)
    plt.show()
    
    plt.plot([*range(len(pca.explained_variance_ratio_))], pca.explained_variance_ratio_, 'bx-')
    plt.xlabel('Explained Variance Ratio')
    plt.ylabel('Variance')
    plt.title('PCA')
    plt.show()
```
