```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read csv input file
X = pd.read_csv("combined_data_no-states_no-header.csv", sep=",")
```

```python
pca = PCA()
pca.fit(X)
print(pca.singular_values_)
```

```python
plt.plot([*range(len(pca.singular_values_))][1:], pca.singular_values_[1:], 'bx-')
plt.xlabel('Singular Values')
plt.ylabel('Variance')
plt.title('PCA')
plt.show()
```

```python
plt.plot([*range(len(pca.explained_variance_ratio_))][1:], pca.explained_variance_ratio_[1:], 'bx-')
plt.xlabel('Explained Variance Ratio')
plt.ylabel('Variance')
plt.title('PCA')
plt.show()
```
